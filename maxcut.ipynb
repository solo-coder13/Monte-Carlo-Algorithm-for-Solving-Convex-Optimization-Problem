```python
import numpy as np

def generate_random_feasible_solution(n):
    return np.random.uniform(-1, 1, size=n)

def is_feasible(solution):
    return np.all(solution >= -1) and np.all(solution <= 1)

def monte_carlo_sampling(num_samples, num_vertices):
    feasible_samples = []
    num_generated = 0
    while len(feasible_samples) < num_samples:
        solution = generate_random_feasible_solution(num_vertices)
        if is_feasible(solution):
            feasible_samples.append(solution)
        num_generated += 1
    return feasible_samples

# Example usage
num_samples = 10
num_vertices = 5
samples = monte_carlo_sampling(num_samples, num_vertices)
print("Generated feasible samples:")
for sample in samples:
    print(sample)
```

```python
import numpy as np
from scipy.io import mmread

# Function to generate a random feasible solution
def generate_random_solution(num_vertices):
    return np.random.uniform(-1, 1, size=num_vertices)

# Function to check if a solution is feasible
def is_feasible(solution):
    return np.all(solution >= -1) and np.all(solution <= 1)

# Load graph from .mtx file
file_path = "/content/MANN-a27.mtx"
graph = mmread(file_path).toarray()

# Number of vertices
num_vertices = len(graph)

# Monte Carlo sampling
num_samples = 10
feasible_samples = []
while len(feasible_samples) < num_samples:
    solution = generate_random_solution(num_vertices)
    if is_feasible(solution):
        feasible_samples.append(solution)

# Print generated feasible samples
for sample in feasible_samples:
    print(sample)
```
